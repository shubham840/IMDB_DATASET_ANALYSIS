```python
# FIRST WE NEED TO SEPERATE THE GENERS COLUMNS VALUE USING THE PIPE(|) 
```

```python
Tempgenre=data.genres.str.split('|',expand=True).iloc[:,0:2]

# data.genres.str-> converted the values to string
#.split('|') now splitted the values with the pipe 
# expand=True it basicaaly expand the splitted data into columns
# .iloc[:,0:2] this iloc method is basically accessing the all rows but 
# ONLY FIRST 2 COLUMNS
```

```python
Tempgenre
```

```python
# NOW GIVING THE COLUMNS NAME OF Tempgenre dataframe
```

```python
Tempgenre.columns=['genre_1','genre_2']
```

```python
Tempgenre
```

```python
# NOW REPLACING THE MISSING VALUES of GENRE_2 WITH VALUES of GENRE_1
```

```python
Tempgenre=Tempgenre.fillna(method='ffill',axis=1)

# method=ffill means it will fill the values in the forward direction
# axis=1 means we have to fill the present column value in the
# next column null value
```

```python
Tempgenre.isnull().sum()

# SO NO NULL VALUES ALL NULL VALUES ARE REPLACED
```

```python
# NOW ADDING THIS TEMOGENRE DATA FRAME TO THE ORIGINAL MOVIES DATA FRAME
```

```python
# FOR THIS WE WILL USE CONCAT FUNCTION 
```

```python
data=pd.concat([data,Tempgenre],axis=1)

# here we are concatinating the tempgenre data with the original movies data
# along the columns
```
